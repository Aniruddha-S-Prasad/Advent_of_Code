```python
import numpy as np
```

```python
with open('input.txt', 'r') as height_map_file:
    height_map =[]
    for line in height_map_file:
        line = line.strip()
        row = []
        for height in line:
            row.append(int(height))
        height_map.append(row)

height_map = np.array(height_map)
visible_map = np.full(height_map.shape, 0)
```

```python
def senic_score_calculator(height_map: np.ndarray, location: tuple[int, int]):
    row, col = location
    tree_house_height = height_map[row, col]
    # Tree heights as seen by the proposed tree house at location
    left_tree_heights = np.flip(height_map[row, :col])
    right_tree_heights = height_map[row, col+1:]
    top_tree_heights = np.flip(height_map[:row, col])
    bottom_tree_heights = height_map[row+1:, col]

    ordered_tree_heights = [top_tree_heights, right_tree_heights, bottom_tree_heights, left_tree_heights]
    score = 1

    for tree_heights in ordered_tree_heights:
        tree_heights: np.ndarray
        view = 0
        for tree_height in tree_heights:
            if tree_height < tree_house_height:
                view += 1
            else:
                break
        if view != tree_heights.shape[0]:
            view += 1
        score = score * view
    # print(height_map)
    # print(f'{top_tree_heights}\n{right_tree_heights}\n{bottom_tree_heights}\n{left_tree_heights}')
    # print(scores)
    # print(np.prod(scores))
    return score
```

```python
senic_scores = np.zeros(height_map.shape, dtype=np.int32)
for index in np.ndindex(height_map.shape):
    if index[0]*index[1] == 0 or index[0] == height_map.shape[0]-1 or index[1] == height_map.shape[1]-1:
        continue
    senic_scores[index] = senic_score_calculator(height_map, index)

print(senic_scores.max())
```
